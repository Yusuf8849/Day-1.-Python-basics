```python
print("===Temperature Converter===")
print("1.Convert Celcius to Fahrenheit")
print("2.Convert Fahrenheit to Celcius")
```

```python
choice = input("Enter your choice (1 or 2):")
```

```python
if choice == "1":
    celcius = float(input("Enter your temperature in Celcius:"))
    fahrenheit = (clecius * 9/5) + 32
    print(f"\n{celcius}°C is equal to {fahrenheit:.1f}°F")
elif choice == "2":
    fahrenheit = float(input("Enter your temperature in Fahrenheit:"))
    celcius = (fahrenheit - 32) * 5/9
    print(f"\n{fahrenheit}°F is equal to {celcius:.1f}°C")
else:
    print("\nInvalid choice!Please restart and Enter either 1 or 2.")
    
```
